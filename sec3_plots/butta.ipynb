```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

# --- 1. Generate Sample Data ---
# Create some random data for the scatter plot
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.randint(0, 3, 50)  # Three different color categories

# --- 2. Create the Scatter Plot ---
fig, ax = plt.subplots()

scatter = ax.scatter(x,
                     y,
                     c=colors,                # Color of the points
                     facecolors='white',      # Set face color to white
                     edgecolors='black',      # Set edge color to black
                     linewidths=1.5,
                     s=100)                   # Size of the markers

# --- 3. Create the First Legend (for Colors) ---
# The handles and labels can be retrieved from the scatter plot object
color_legend = ax.legend(*scatter.legend_elements(),
                         loc="upper right",
                         title="Color Meaning")

# --- 4. Add the First Legend to the Axes ---
# This is crucial for having two legends
ax.add_artist(color_legend)

# --- 5. Create the Second Legend (for Markers) ---
# Create proxy artists for the marker legend
# These are dummy lines with the desired marker style but no data
marker_legend_handles = [Line2D([0], [0],
                                marker='o',
                                color='w',
                                label='Data Point',
                                markerfacecolor='white',
                                markeredgecolor='black',
                                markersize=10)]

# Create the second legend using these proxy artists
ax.legend(handles=marker_legend_handles, loc='lower right', title="Marker Style")


# --- 6. Final Touches ---
ax.set_title("Scatter Plot with Two Legends")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```
